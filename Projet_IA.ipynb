```python
def structured_extract_details(div):
    details = {}

    # Extract Marque and Model
    marque_model = div.find('h2')
    details["Marque et Modèle"] = marque_model.get_text(strip=True) if marque_model else None

    # Extract Version
    version = marque_model.find_next_sibling('div') if marque_model else None
    details["Version"] = version.get_text(strip=True) if version else None

    # Extract other characteristics
    characteristics_div = div.find('div', class_='Vehiculecard_Vehiculecard_characteristics')
    if characteristics_div:
        characteristics = characteristics_div.find_all('div', class_='Text_Text_text')
        details["Année"] = characteristics[0].get_text(strip=True) if len(characteristics) > 0 else None
        details["Kilométrage"] = characteristics[1].get_text(strip=True) if len(characteristics) > 1 else None
        details["Type de Transmission"] = characteristics[2].get_text(strip=True) if len(characteristics) > 2 else None
        details["Type de Carburant"] = characteristics[3].get_text(strip=True) if len(characteristics) > 3 else None

    # Extract price
    price = div.find('span', class_='Vehiculecard_Vehiculecard_price')
    details["Prix"] = price.get_text(strip=True) if price else None

    # The remaining details (Evaluation, Warranty, Delivery) will be extracted based on text as they are not structured uniformly in separate tags
    remaining = div.get_text(strip=True)
    details["Évaluation du Prix"] = re.search(r'(Bonne affaire|Offre équitable|Analyse indisponible)', remaining).group(1) if re.search(r'(Bonne affaire|Offre équitable|Analyse indisponible)', remaining) else None
    details["Garantie"] = re.search(r'(Garantie \d+ mois)', remaining).group(1) if re.search(r'(Garantie \d+ mois)', remaining) else None
    details["Livraison"] = re.search(r'(Livraison\d+)', remaining).group(1) if re.search(r'(Livraison\d+)', remaining) else None

    return details

# Extract details from the divs using the structured approach and create the dataset
df_structured = pd.DataFrame([structured_extract_details(div) for div in divs])
df_structured
```

```python
import pandas as pd
import time
import re

# Function to extract vehicle details from a div
def structured_extract_details(div):
    details = {}

    # Extract Marque and Model
    marque_model = div.find('h2')
    details["Marque et Modèle"] = marque_model.get_text(strip=True) if marque_model else None

    # Extract Version
    version = marque_model.find_next_sibling('div') if marque_model else None
    details["Version"] = version.get_text(strip=True) if version else None

    # Extract other characteristics
    characteristics_div = div.find('div', class_='Vehiculecard_Vehiculecard_characteristics')
    if characteristics_div:
        characteristics = characteristics_div.find_all('div', class_='Text_Text_text')
        details["Année"] = characteristics[0].get_text(strip=True) if len(characteristics) > 0 else None
        details["Kilométrage"] = characteristics[1].get_text(strip=True) if len(characteristics) > 1 else None
        details["Type de Transmission"] = characteristics[2].get_text(strip=True) if len(characteristics) > 2 else None
        details["Type de Carburant"] = characteristics[3].get_text(strip=True) if len(characteristics) > 3 else None

    # Extract price
    price = div.find('span', class_='Vehiculecard_Vehiculecard_price')
    details["Prix"] = price.get_text(strip=True) if price else None

    # The remaining details (Evaluation, Warranty, Delivery) will be extracted based on text as they are not structured uniformly in separate tags
    remaining = div.get_text(strip=True)
    details["Évaluation du Prix"] = re.search(r'(Bonne affaire|Offre équitable|Analyse indisponible)', remaining).group(1) if re.search(r'(Bonne affaire|Offre équitable|Analyse indisponible)', remaining) else None
    details["Garantie"] = re.search(r'(Garantie \d+ mois)', remaining).group(1) if re.search(r'(Garantie \d+ mois)', remaining) else None
    details["Livraison"] = re.search(r'(Livraison\d+)', remaining).group(1) if re.search(r'(Livraison\d+)', remaining) else None

    return details

# List to store all vehicle details
all_vehicle_details = []

# Parcourir chaque page
page_numbers = range(1, 300)  # Assuming there are 9 pages
for page_number in page_numbers:
    url = URL_BASE + PAGE_STRUCTURE + str(page_number)
    time.sleep(3)  # Pause for 1 second between each request
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract all divs containing vehicle details
        divs = soup.find_all('div', class_='Vehiculecard_Vehiculecard_cardBody')

        # Extract details for each div and add to the list
        for div in divs:
            all_vehicle_details.append(structured_extract_details(div))
    else:
        print(f"Erreur lors de l'accès à l'URL : {url}")

# Convert the list of vehicle details to a DataFrame
df_structured = pd.DataFrame(all_vehicle_details)
df_structured
```

# Modèles de ML

```python
# Séparer les caractéristiques (X) et la variable cible (y)
X = data[['Marque et Modèle', 'Version', 'Année', 'Kilométrage', 'Type de Transmission', 'Type de Carburant']]
y = data['Prix']

# One-Hot Encoding pour les variables catégorielles
X_encoded = pd.get_dummies(X, columns=['Marque et Modèle', 'Version', 'Type de Transmission', 'Type de Carburant'], drop_first=True)

# Séparer les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Créer le modèle de régression linéaire
model = LinearRegression()

# Entraîner le modèle sur l'ensemble d'entraînement
model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluer les performances du modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Afficher les performances du modèle
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")

# Vous pouvez également afficher les coefficients du modèle
print("Coefficients du modèle :")
for feature, coef in zip(X_train.columns, model.coef_):
    print(f"{feature}: {coef}")
```

# Deep Learning Model

```python
# Séparer les caractéristiques (X) et la variable cible (y)
X = data[['Marque et Modèle', 'Version', 'Année', 'Kilométrage', 'Type de Transmission', 'Type de Carburant']]
y = data['Prix']

# One-Hot Encoding pour les variables catégorielles
X_encoded = pd.get_dummies(X, columns=['Marque et Modèle', 'Version', 'Type de Transmission', 'Type de Carburant'], drop_first=True)

# Standardisation des données - Les réseaux de neurones sont sensibles à la mise à l'échelle des caractéristiques
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Séparer les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialisation du modèle
model = Sequential()

# Ajouter des couches au modèle
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # Couche de sortie pour la régression

# Compiler le modèle
model.compile(optimizer='adam', loss='mean_squared_error')

# Entraîner le modèle
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))


# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluer les performances du modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Afficher les performances du modèle
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

## Modèle final sélectionné: RandomForestRegressor

```python
# Séparer les caractéristiques (X) et la variable cible (y)
X = data[['Marque et Modèle', 'Version', 'Année', 'Kilométrage', 'Type de Transmission', 'Type de Carburant']]
y = data['Prix']

# One-Hot Encoding pour les variables catégorielles
X_encoded = pd.get_dummies(X, columns=['Marque et Modèle', 'Version', 'Type de Transmission', 'Type de Carburant'], drop_first=True)

# Séparer les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Créer le modèle RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)  # n_estimators définit le nombre d'arbres

# Entraîner le modèle sur l'ensemble d'entraînement
model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluer les performances du modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Afficher les performances du modèle
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")

# Afficher l'importance des caractéristiques si nécessaire
print("Importance des caractéristiques :")
for feature, importance in zip(X_train.columns, model.feature_importances_):
    print(f"{feature}: {importance}")
```
